## 2. Explore Output Variable (quality): Majority of wines are of normal quality, few are extremely poor, while few are excellent, which means the data is imbalanced. It's better to add a bucket variable as output variable.

```python
fig, ax = plt.subplots(1,1,figsize = (10,6))
sns.countplot(data = df, x = 'quality',color = '#107ab0')
ax.set_title('Distribution of Wine Quality', fontsize = 16)
plt.show()
```

## 3. Explore Features Distribution

### 3.1 Distrubution of Categorical features: There were more white wines than red and they were roughly in 3 to 1 ratio.

```python
from matplotlib import cm
fig, axes = plt.subplots(1,2,figsize = (12,7))

# Draw bar chart
sns.countplot(data = df,x='type',palette = 'Blues_r',ax = axes[0])

# Draw pie chart
colors = cm.Blues_r(np.linspace(.3,.7, 2))
names = df['type'].value_counts().index.tolist()
size = df['type'].value_counts().values.tolist()
plt.pie(size, labels = names, colors=colors,autopct = '%1.1f%%',startangle = 90)

# Set title
fig.suptitle("Red & White Wines Count and Share", fontsize = 16)
plt.show()
```

### 3.2 Check Quality difference between white and red wine: Average quality of White wines is bigger then White wines, and also white wines has bigger max quality value.

```python
quality = df.groupby('type').agg(['mean','std','count','min','max'])['quality']
quality = quality.reset_index()
f, ax = plt.subplots(1,2,figsize=(12,7))
sns.barplot(data=quality,y='mean',x='type',palette='Blues_r',saturation=0.9,ax=ax[0])
sns.boxplot(data=df,y='quality',linewidth=2.5,x='type',palette='Blues_r',ax=ax[1])
f.suptitle("Red & White Wines Quality (Mean)", fontsize=16)
```

### 3.3 Distribution of Numerical features: pH feature is approximately normally distributed; the other features are positively skewed, it means that there is more weight in the left tail of the distribution. 

```python
df2 = df.iloc[:, 1:13]
plt.figure(figsize = (20, 16))
for i, col in enumerate(list(df2.columns.values)):
    plt.subplot(4,3,i+1)
    sns.distplot(df2[col], color = '#107ab0', kde = 'True', label = 'data')
```

## 4. Check outliers of variables: all the features shows outliers.

```python
plt.figure(figsize = (16,20))
for i, col in enumerate(list(df2.columns.values)):
    plt.subplot(4,3,i+1)
    sns.boxplot(y=df2[col],palette='Blues_r',width=0.3)
```

## 5. Visualizing Correlation Matrix of Attributes
Observations:
1. Wine quality is highly correlated with alcohol quantity and density. However, alcohol and density are negatively correlated. Therefore, one of them can be used as wine quality predictor. 
2. Wine quality is negatively correlated with the volatile acidity
3. pH is negatively correlated with fixed acidity, citric acid, total SO2 and residual sugar. 

```python
corr = df2.corr()
        
f,ax = plt.subplots(figsize=(12,9))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr,cmap='YlGnBu',annot=True,linewidths=.1,square=True,mask=mask);
plt.title('Correlation map for Variables',fontsize=15);
```
